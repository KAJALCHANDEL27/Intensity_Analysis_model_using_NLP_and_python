# 2. Text preprocessing: 
In this step, we will clean the ‚Äòcontent‚Äô column. We will remove the unwanted HTML tags, brackets, or special characters that may be present in the texts. We will use Regex to clean the text.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re

# Downloading required nltk data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')  # Needed for lemmatization
nltk.download('omw-1.4')  # Additional resources for lemmatization

# Stop words set
stop_words = set(stopwords.words('english'))

# Initialize stemmer and lemmatizer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to clean, tokenize, stem, and lemmatize content
def clean_tokenize_stem_lemmatize(content):
    # Removing html brackets and other square brackets from the string using regex
    content = re.sub(r'<.*?>', '', content)

    # Removing special characters like @, #, $, etc
    content = re.sub(r'[^a-zA-Z0-9\s]', '', content)

    # Removing numbers
    content = re.sub(r'\d+', '', content)

    # Converting text to lower case
    content = content.lower()

    # Tokenization of words
    tokens = word_tokenize(content)

    # Stop words removal
    tokens = [word for word in tokens if word not in stop_words]

    # Apply stemming and lemmatization
    stemmed = [ps.stem(word) for word in tokens]
    lemmatized = [lemmatizer.lemmatize(word) for word in stemmed]

    return lemmatized  # Returning a list of lemmatized tokens

# Applying the function on the dataset
intensity_df['tokenized_content'] = intensity_df['content'].apply(clean_tokenize_stem_lemmatize)

# Verify changes (original content and tokenized version)
print(intensity_df[['content', 'tokenized_content']].head())
```

# Step 5. Feature Engineering

a. Textual Features:



# Combine TF-IDF and N-gram Features
1:TF-IDF Vectorization: As you've already done, we will first create TF-IDF features from the cleaned text.

2:N-gram Features: Next, we will create n-gram features using CountVectorizer.

3:Combining Features: Finally, we will combine both feature matrices into one.


```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from scipy.sparse import hstack

# Step 1: Convert the tokenized content back to a string for TF-IDF and N-grams
intensity_df['joined_content'] = intensity_df['tokenized_content'].apply(lambda x: ' '.join(x))

# Step 2: Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # Limit to 5000 most common words
tfidf_features = tfidf_vectorizer.fit_transform(intensity_df['joined_content'])

# Step 3: Initialize CountVectorizer for n-grams (e.g., unigrams, bigrams, and trigrams)
ngram_vectorizer = CountVectorizer(ngram_range=(1, 3), max_features=5000)  # Adjust n-grams as needed
ngram_features = ngram_vectorizer.fit_transform(intensity_df['joined_content'])

# Step 4: Combine TF-IDF and N-gram features
combined_features = hstack((tfidf_features, ngram_features))

# Optional: Convert to a dense matrix (more memory-intensive)
combined_dense = combined_features.toarray()

# Print the shape of the combined features
print("Combined features shape:", combined_dense.shape)

# Optional: Check some feature names for verification
feature_names = tfidf_vectorizer.get_feature_names_out()
ngram_names = ngram_vectorizer.get_feature_names_out()

# Print some of the feature names from TF-IDF and n-grams
print("Sample TF-IDF feature names:", feature_names[:10])  # Adjust index for more/less
print("Sample n-gram feature names:", ngram_names[:10])  # Adjust index for more/less
```

This combined feature set will provide a comprehensive representation of your text data, capturing both the significance of individual words (TF-IDF) and the context provided by n-grams. You can now use this combined feature set for training your machine learning model.









# Part-of-Speech (POS) Tags: Capture grammatical information.

To incorporate Part-of-Speech (POS) tagging into your feature engineering process, you can use the nltk library, which provides functionality for POS tagging in Python. Below, I'll demonstrate how to add POS tags as features to your existing DataFrame. This will involve creating a new column with the POS tags for each tokenized content.

Step 1: Import Necessary NLTK Modules
Make sure you have NLTK's POS tagger available. You may need to download the necessary datasets if you haven‚Äôt already:

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

Step 2: Define a Function to Generate POS Tags
You can create a function that applies POS tagging to the list of tokens:



```python
# Function to get POS tags for a list of tokens
def get_pos_tags(tokens):
    return nltk.pos_tag(tokens)  # Returns a list of tuples (word, POS tag)

# Apply the function to the tokenized content
intensity_df['pos_tags'] = intensity_df['tokenized_content'].apply(get_pos_tags)
```

Step 3: Extract POS Tag Features
You might want to create numerical features based on the POS tags. For instance, you can count the occurrence of each POS tag:



```python
# Create a function to count POS tags
def pos_tag_counts(pos_tags):
    pos_counts = {}
    for word, pos in pos_tags:
        if pos in pos_counts:
            pos_counts[pos] += 1
        else:
            pos_counts[pos] = 1
    return pos_counts

# Apply the function to create a new DataFrame for POS tag counts
pos_counts_df = intensity_df['pos_tags'].apply(pos_tag_counts).apply(pd.Series).fillna(0)

# Combine the POS counts with the original DataFrame
intensity_df = pd.concat([intensity_df, pos_counts_df], axis=1)
```

Step 4: Verify the New Features
You can check the updated DataFrame to verify that the POS tags and their counts have been added:


```python
# Print the first few rows of the updated DataFrame
print(intensity_df[['content', 'tokenized_content', 'pos_tags'] + list(pos_counts_df.columns)].head())
```

This approach enriches your dataset with grammatical information, which can be useful for understanding the structure of the text and may improve the performance of your machine learning models.

#  Lexical Features
You can create features to count exclamations, all-caps words, smileys, etc. This can provide additional insights into the emotional tone of the text.

Step 1: Define Lexical Feature Functions


```python
import re

# Function to count exclamations
def count_exclamations(text):
    return text.count('!')

# Function to count all-caps words
def count_all_caps(text):
    return len(re.findall(r'\b[A-Z]{2,}\b', text))

# Function to count smileys
def count_smileys(text):
    smiley_pattern = r'(:\)|:\(|:\d|;\)|;\()'  # Add more patterns as needed
    return len(re.findall(smiley_pattern, text))

# Apply lexical feature functions to create new columns
intensity_df['exclamation_count'] = intensity_df['content'].apply(count_exclamations)
intensity_df['all_caps_count'] = intensity_df['content'].apply(count_all_caps)
intensity_df['smiley_count'] = intensity_df['content'].apply(count_smileys)
```

This comprehensive feature engineering approach will enhance your dataset by capturing various aspects of the text, which can improve the performance of your machine learning models.








# Padding
The length of each sentence is different. To pass it through the model, the length of each sentence should be equal. By visualizing the dataset, we can see that the length of the sentence in the dataset is not greater than 100 words. So, now we will convert every sentence to 100 words. For this, we will take the help of padding.



```python
from tensorflow.keras.preprocessing.sequence import pad_sequences  # Make sure this is imported
from tensorflow.keras.preprocessing.text import Tokenizer

# Tokenizer instance to convert words into integers
tokenizer = Tokenizer()
tokenizer.fit_on_texts(intensity_df['tokenized_content'])

# Convert text (tokenized words) to sequences of integers
sequences = tokenizer.texts_to_sequences(intensity_df['tokenized_content'])  # This step assigns 'sequences'

# Padding the sequences to make all tokenized sentences of length 100
# (using a max length of 100)
padded_sequences = pad_sequences(sequences, maxlen=100, padding='post')

# Adding the padded sequences back to the dataframe as a new column
intensity_df['padded_content'] = list(padded_sequences)

# Verify changes (original tokenized content, sequence version, and padded version)
print(intensity_df[['tokenized_content', 'padded_content']].head())
```

# Step 6: Model Selection / Model Training / Model Evaluation / Hyperparameter Tuning
a. Traditional Algorithms:
Logistic Regression are good starting models. Here's an example using Logistic Regression:

b.Train the selected models using the preprocessed data.

c.Assess the performance of the trained models using appropriate evaluation metrics.

d.Optimize model hyperparameters to improve predictive accuracy using GridSearchCV


```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
import seaborn as sns
import matplotlib.pyplot as plt

# Features and labels
X = combined_dense  # Features
y = intensity_df['intensity']  # Labels

# Set the random seed for reproducibility
np.random.seed(42)

# Generate a random permutation of indices based on the length of the data
indices = np.random.permutation(len(X))

# Set the train-test split ratio (e.g., 80% train, 20% test)
split_ratio = 0.8
train_size = int(split_ratio * len(X))

# Split the indices for training and testing
train_indices = indices[:train_size]
test_indices = indices[train_size:]

# Use the indices to split the data into training and test sets
X_train, X_test = X[train_indices], X[test_indices]
y_train, y_test = y_encoded[train_indices], y_encoded[test_indices]

# Output the shapes of the datasets
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")


# Define the Logistic Regression model
logistic_model = LogisticRegression()

# Define the hyperparameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Inverse of regularization strength
    'penalty': ['l1', 'l2'],        # Regularization type
    'solver': ['liblinear', 'saga'] # Algorithm to use in the optimization problem
}

# Set up Grid Search
grid_search = GridSearchCV(estimator=logistic_model, param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1)

# Fit the model with Grid Search
grid_search.fit(X_train, y_train)

# Get the best parameters and score
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation accuracy: ", grid_search.best_score_)

# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Convert predictions back to original labels
try:
    y_pred_labels = label_encoder.inverse_transform(y_pred)
except ValueError as e:
    print("Encountered unseen labels during prediction:", e)

# Optional: Mapping unseen labels to a default value
known_labels = set(label_encoder.classes_)
y_pred_mapped = ['Unknown' if label not in known_labels else label for label in y_pred_labels]

print("Mapped predicted labels:", np.unique(y_pred_mapped))


# Predictions and evaluation
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Visualize the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()

# Save the confusion matrix heatmap
plt.savefig('visuals/confusion_matrix_logistic_regression.png', format='png')
plt.show()

```

```python
#Saving Feature Importance Plot

# Feature importance using logistic regression coefficients
importance = best_model.coef_[0]  # Get coefficients for the best model
feature_names = np.array([f'Feature {i+1}' for i in range(X.shape[1])])  # Replace with actual feature names if available
indices = np.argsort(importance)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(feature_names[indices], importance[indices], color='blue')
plt.xlabel('Coefficient Value')
plt.title('Feature Importance from Logistic Regression')
plt.savefig('visuals/feature_importance_logistic_regression.png', format='png')
plt.show()
```

# Overview of pickle
pickle is a Python module used for serializing and deserializing Python objects. Serialization refers to the process of converting a Python object into a byte stream (saving it), and deserialization is the reverse process (loading it back into memory). This is particularly useful in machine learning for saving trained models and any associated preprocessing steps.

# Serialization refers to the process of converting a Python object into a byte stream (saving it)

1. Saving Objects


```python
import pickle

# Save the best model
with open('logistic_model.pkl', 'wb') as model_file:
    pickle.dump(best_model, model_file)

# Save the TF-IDF vectorizer (if you created it separately)
with open('tfidf_vectorizer.pkl', 'wb') as vectorizer_file:
    pickle.dump(tfidf_vectorizer, vectorizer_file)

# Save the CountVectorizer (if you created it separately for n-grams)
with open('ngram_vectorizer.pkl', 'wb') as ngram_file:
    pickle.dump(ngram_vectorizer, ngram_file)

# Optionally save other relevant objects, like LabelEncoder
with open('label_encoder.pkl', 'wb') as le_file:
    pickle.dump(label_encoder, le_file)

# Optionally save the combined features if you need it later
with open('combined_features.pkl', 'wb') as features_file:
    pickle.dump(combined_dense, features_file)
```
